```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
sss = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=23)
```

```python
sss.get_n_splits(X, y)
```

```python
for train_index, test_index in sss.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

## Key Ideas for K-Nearest Neighbors

1. K-Nerest Neighbors (KNN) classifies a record by assigning it to the class that similar records belong to. 

2. Similarity (distance) is determined by Euclidian distance or other related metrics 

3. The number of nearest neighbors to compare a record to, K, is determined by how well the algorithm performs on training data, using different values for K.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc_model = KNeighborsClassifier().fit(X_train, y_train)
```

```python
prediction = knc_model.predict_proba(X_test)
```

```python
from sklearn.metrics import log_loss
```

```python
print(log_loss(y_test, prediction))
```

```python
knn = []
log_loss_ = []

for i in range(1,16):
    knc_model = KNeighborsClassifier(n_neighbors=i).fit(X_train, y_train)
    predictions = knc_model.predict_proba(X_test)
    log_loss_.append(log_loss(y_test, predictions))
    knn.append(i)
```

```python
plt.plot(range(1,16), log_loss_)
plt.xlabel('Number of Nnearest Neighbours')
plt.ylabel('Log Loss')
plt.title('KNN Elbow Method')
```
